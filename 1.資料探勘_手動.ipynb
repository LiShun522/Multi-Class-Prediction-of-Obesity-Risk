```python
# id	編號	每筆資料的唯一識別碼
# Gender	性別	受測者的性別（男/女）
# Age	年齡	受測者的年齡（單位：歲）
# Height	身高	受測者的身高（單位：公尺）
# Weight	體重	受測者的體重（單位：公斤）
# family_history_with_overweight	家族肥胖史	受測者的家族是否有肥胖病史（是/否）
# FAVC	高熱量食物攝取頻率	受測者是否經常攝取高熱量食物（是/否）
# FCVC	蔬菜攝取頻率	受測者的蔬菜攝取頻率（數值越高表示吃得越頻繁）
# NCP	每日餐次	受測者每天進食的餐數（例如：2、3、4）
# CAEC	進食零食頻率	受測者是否經常吃零食（從不、偶爾、經常、總是）
# SMOKE	是否吸菸	受測者是否有吸菸習慣（是/否）
# CH2O	每日飲水量	受測者每天攝取的水量（數值越高表示喝得越多）
# SCC	是否有慢性疾病	受測者是否有慢性疾病（是/否）
# FAF	每週運動頻率	受測者每週運動的頻率（數值越高表示運動次數越多）
# TUE	使用電子產品時間	受測者每天花多少時間使用電子產品（如手機、電腦，數值越高表示時間越長）
# CALC	飲酒頻率	受測者的酒精攝取頻率（從不、偶爾、經常、總是）
# MTRANS	主要交通工具	受測者日常使用的主要交通方式（步行、腳踏車、汽車、機車、大眾運輸等）
# NObeyesdad	肥胖風險分類	受測者的肥胖風險類別（多分類：如健康、過重、輕度肥胖、中度肥胖、重度肥胖等）
```

```python
# 將 Gender改變成數值型，丟掉第一筆，保留一組 bool就好
data = pd.get_dummies(data, columns= ["Gender"], drop_first=True)
```

```python
# 有大小分布，適合 特徵縮放 Standard Scaling 
# "Age", "Height", "Weight"
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
data[["Age", "Height", "Weight"]] = std_scaler.fit_transform(data[["Age", "Height", "Weight"]])
```

```python
# 0/1類型使用.map()方式
# "family_history_with_overweight", "FAVC", "SMOKE", "SCC" 
# 將這些欄位先設立一個 yes/no 的 mapping
yes_no_mapping = {"yes":1, "no":0}
data["family_history_with_overweight"] = data["family_history_with_overweight"].map(yes_no_mapping)
data["FAVC"] = data["FAVC"].map(yes_no_mapping)
data["SMOKE"] = data["SMOKE"].map(yes_no_mapping)
data["SCC"] = data["SCC"].map(yes_no_mapping)
```

```python
# 有序適合 Label Encoding ，使用.map()方式
# "CAEC", "CALC", "NObeyesdad"
CAEC_mapping = {"no":0, "Sometimes":1, "Frequently":2, "Always":3}
# data["CALC"].unique() 只有3個不同，測試資料 test.csv中，data["CALC"]有4個不同，要將'Always'改為'Frequently'
CALC_mapping = {"no":0, "Sometimes":1, "Frequently":2, "Always":2}
NObeyesdad_mapping = {
    "Insufficient_Weight": 0, 
    "Normal_Weight": 1,   
    "Overweight_Level_I": 2,   
    "Overweight_Level_II": 3, 
    "Obesity_Type_I": 4, 
    "Obesity_Type_II": 5, 
    "Obesity_Type_III": 6 
}
data["CAEC"] = data["CAEC"].map(CAEC_mapping)
data["CALC"] = data["CALC"].map(CAEC_mapping)
data["NObeyesdad"]= data["NObeyesdad"].map(NObeyesdad_mapping)
```

```python
# 老師自寫 One-Hot Encoding
def one_hot_encode(data, field_name):
	field_value = data[field_name].value_counts().index
	new_field_name = []
	for value in field_value:
		fn = field_name+"_"+value
		data[fn] = (data[field_name] == value).astype('int64')
		new_field_name.append(fn)
	return new_field_name
```

```python
# 計算與 NObeyesdad相關係數最高的 Top 10
```

```python
# 定義目標欄位名稱
target_col = "NObeyesdad"  # 或 y_encoded 如果已將其添加為 data 的一列
# 計算所有特徵與目標欄位的相關性
other_cols = [col for col in data.columns if col != target_col]
correlations = []
for col in other_cols:
    # 計算該特徵與目標變數的相關性(絕對值)
    # 使用 pandas corr 函數計算相關係數
    correlation = abs(data[col].corr(data[target_col]))
    correlations.append((col, correlation))
# 排序結果
top_features = pd.DataFrame(correlations, columns=["Feature", "Correlation"]).sort_values("Correlation", ascending=False)
# 顯示前10個最相關的特徵
print("與肥胖程度最相關的前10個特徵:")
print(top_features.head(10))

# 使用seaborn繪製條形圖
try:
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    # 設置seaborn風格
    sns.set(style="whitegrid")
    
    # 繪製前10名特徵的相關性
    plt.figure(figsize=(12, 8))
    
    # 修正 barplot - 使用 hue 參數並設置 legend=False
    ax = sns.barplot(
        x="Correlation", 
        y="Feature", 
        hue="Feature",  # 新增 hue 參數
        data=top_features.head(10),
        palette="viridis",
        legend=False  # 不顯示圖例
    )
    
    # 添加標籤和標題
    plt.xlabel("Correlation Coefficient (Absolute Value)", fontsize=12)
    plt.ylabel("Feature Name", fontsize=12)
    plt.title("Top 10 Features Correlated with Obesity Level", fontsize=14)
    
    # 在條形上添加數值標籤
    for i, v in enumerate(top_features.head(10)["Correlation"]):
        ax.text(v + 0.01, i, f"{v:.3f}", va='center')
    
    plt.tight_layout()
    plt.show()
    
    # 可選：繪製熱力圖
    plt.figure(figsize=(12, 10))
    
    # 選擇前10個特徵加上目標變數
    top_features_names = top_features.head(10)["Feature"].tolist()
    selected_cols = top_features_names + [target_col]
    
    # 計算相關矩陣
    corr_matrix = data[selected_cols].corr()
    
    # 使用seaborn的熱力圖
    sns.heatmap(
        corr_matrix, 
        annot=True, 
        cmap="coolwarm", 
        linewidths=0.5, 
        fmt=".2f",
        vmin=-1, 
        vmax=1
    )
    
    plt.title("Correlation Heatmap of Top Features with Obesity", fontsize=14)
    plt.tight_layout()
    plt.show()
    
except Exception as e:
    print(f"繪圖時發生錯誤: {e}")
```
